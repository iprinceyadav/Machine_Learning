# Matplotlib library learning


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
x=np.linspace(1,10,100)
```

```python
y=np.sin(x)
```

```python
z=np.cos(x)
```

```python
print(x)
```

## Ploting the data

```python
#sinwave 
```

```python
plt.plot(x,y)
plt.show()
```

```python
#COSINE 
plt.plot(x,z)
plt.show()
```

## adding title x-axis and y -axis label

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

## parabola

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)

plt.ylabel('Y')
plt.xlabel('X')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,'g*')
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
```

## BAR PLOT

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=["English","french","spanish","german","latin"]
people=[100,50,150,40,80]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('No of people')
plt.show()
```

# Pie chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=["English","french","spanish","german","latin"]
people=[100,50,150,40,80]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

## Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

## 3D ScatterPlot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
